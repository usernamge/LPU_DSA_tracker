#### 13. Roman to Integer

- https://leetcode.com/problems/roman-to-integer/description/?envType=problem-list-v2&envId=string

```python
# Solution 1

class Solution:
    def romanToInt(self, s: str) -> int:
        translations = {
            "I": 1,
            "V": 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000
        }
        number = 0
        print("before",s)
        s = s.replace("IV", "IIII").replace("IX", "VIIII")
        s = s.replace("XL", "XXXX").replace("XC", "LXXXX")
        s = s.replace("CD", "CCCC").replace("CM", "DCCCC")
        print("after",s)
        for char in s:
            number += translations[char]
        return number
    
object = Solution()
object.romanToInt("III")#3
object.romanToInt("LVIII")#58
object.romanToInt("MCMXCIV")#1994

```

### Time Complexity
You perform:
- 6 replace() operations
- 1 pass through the final string
Each replace() scans the string, so:
- 6 × O(n) = O(n)
-  Summation loop = O(n)
Overall: O(n) time.
### Space Complexity
The replace() calls create new strings, so extra space is proportional to the string length:
- O(n) space

```python
# Solution 2

class Solution:
    def romanToInt(self, s: str) -> int:
        values = {
            "I": 1, "V": 5,
            "X": 10, "L": 50,
            "C": 100, "D": 500,
            "M": 1000
        }

        total = 0

        for i in range(len(s) - 1):
            if values[s[i]] < values[s[i + 1]]:
                total -= values[s[i]]
            else:
                total += values[s[i]]

        return total + values[s[-1]]
    
object = Solution()
object.romanToInt("III")#3
object.romanToInt("LVIII")#58
object.romanToInt("MCMXCIV")#1994
```
